```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Tạo dữ liệu mẫu
np.random.seed(42)
group1 = np.random.normal(50, 5, 30)  # Nhóm 1: trung bình 50
group2 = np.random.normal(53, 5, 30)  # Nhóm 2: trung bình 53

# Thực hiện kiểm định t-test hai mẫu độc lập
t_stat, p_value = ttest_ind(group1, group2)

# In kết quả
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value = {p_value:.4f}")

# --- Vẽ biểu đồ histogram ---
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.hist(group1, bins=10, alpha=0.7, label='Group 1 (Mean≈50)')
plt.hist(group2, bins=10, alpha=0.7, label='Group 2 (Mean≈53)')
plt.title("Phân phối của hai nhóm")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# --- Vẽ boxplot ---
plt.subplot(1, 2, 2)
plt.boxplot([group1, group2], tick_labels=['Group 1', 'Group 2'])
plt.title("Biểu đồ Boxplot so sánh hai nhóm")
plt.ylabel("Value")

plt.suptitle(f"Kết quả t-test: p-value = {p_value:.4f}", fontsize=12, color="darkblue")
plt.tight_layout()
plt.show()
```
