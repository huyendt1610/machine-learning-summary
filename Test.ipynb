```python

from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

# Mean vector and covariance matrix
mu = np.array([0, 0])
cov = np.array([[1, 0.8],
                [0.8, 1]])

cdf_matrix = np.array([[1, 0.8],
                [0.8, 1]])

# Create a grid of (x, y)
x, y = np.mgrid[-3:3:.01, -3:3:.01] # tao ra grid 2-D các mảng tọa độ, mỗi D là 1 np.arange(start, stop, step)
pos = np.dstack((x, y)) # Mỗi cặp (x[i,j], y[i,j]) là tọa độ một điểm trên lưới.
# Compute PDF
rv = multivariate_normal(mu, cov)
print(rv.rvs)
z = rv.pdf(pos) #pdf: probability density function (hàm mật độ xác suất) theo cac gia tri (x,y) cua pos
k = rv.cdf(cdf_matrix) # cumulative distribution function (hàm xác suất tích lũy)

k
# Plot contours
plt.contourf(x, y, z, cmap='viridis')
plt.title("Bivariate Normal Distribution")
plt.xlabel("X1")
plt.ylabel("X2")
plt.axis("equal")
plt.show()
```
