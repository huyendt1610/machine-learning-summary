```python
def af(n:int):
    return np.diag(np.array(np.arange(1,n), dtype=int), k = -1)

def myfunc(A:np.array):
    a = A[::2, ::2]
    return np.sum(a)

def myfuncA(A:np.array):
    a = np.min(A, axis=1) 
    b = np.max(A, axis=1)
    return np.sum(b-a)
```
