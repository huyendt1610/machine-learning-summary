```python

coders = []
coders.append(["Eric", "Windows", "Pascal", 10])
coders.append(["Matt", "Linux", "PHP", 2])
coders.append(["Alan", "Linux", "Java", 17])
coders.append(["Emily", "Mac", "Cobol", 9])

print(coders)

with open("abc.txt", "w") as file:
    for line in coders:
        tempLine = ""
        for value in line:
            tempLine += f"{value}," 
        tempLine = tempLine[:-1]
        file.write(f"{tempLine}\n" )
```

```python
import urllib.request
import json

data = urllib.request.urlopen("https://studies.cs.helsinki.fi/stats-mock/api/courses")

courses = json.loads(data.read())
#print(courses)

tulos = []

for c in courses: 
    if(c['enabled'] == True):
        tulos.append((c["fullName"],c['name'], c['year']))

print(tulos)
```

```python
# Excess keyword argument (python 3) example:
def foo(a, b, c, **args): #pack list of key1=value1, key2=value2.. into args
    print("a = %s" % (a,))
    print("b = %s" % (b,))
    print("c = %s" % (c,))
    print(args)

t = {'a': "testa", 'd': "excess", 'c': "testc", 'b': "testb", 'k': "another_excess"}
foo(**t) #unpack dict to list of key1=value1, key2=value2...
```
