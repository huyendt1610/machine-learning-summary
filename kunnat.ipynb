```python
import matplotlib.pyplot as plt
import numpy as np #Numerical python 

```

```python
suurimmatkaupungit = kunnat[kunnat['Väkiluku, 2024'] > 50000]

x = np.array(suurimmatkaupungit['Alue'])
y = np.array(suurimmatkaupungit['Väkiluku, 2024'])

plt.bar(x,y,label="asukasluku")
plt.xticks(rotation=90)
plt.title("Kuva 1: suurimmat kaupungit")
plt.xlabel("kunta")
plt.legend()
plt.show()
```

```python
isoimmatluku =  suurimmatkaupungit['Väkiluku, 2024'].sum()
suurimmatkaupungit['percentage'] = suurimmatkaupungit['Väkiluku, 2024'] / isoimmatluku * 100

#suurimmatkaupungit.loc[:,'percentage']= suurimmatkaupungit['Väkiluku, 2024'] / isoimmatluku * 100
suurimmatkaupungit
```

```python
#13) tee suurimmista kunnista piirasdiagrammi
y = suurimmatkaupungit['percentage']
labels = suurimmatkaupungit['Alue']

plt.pie(y, labels=labels, autopct='%.0f%%')
plt.title('Kuva 2')
plt.show()
#t1122
```

```python
#- tee vielä sekä pylväs-että piirasdiagrammit suurimpien kuntien yhteensä väkiluvun suhteesta muuhun maahan

isoimmmat = isoimmatluku/suomenvakiluku * 100
muut = 100 - isoimmmat
values = [isoimmmat, muut]

kaupungit_count = suurimmatkaupungit['Alue'].count()
label1 = f'isommat {kaupungit_count} kpl'
label2 = f'muut {kunnat['Alue'].count() - kaupungit_count}  kpl'
labels=[label1, label2]

plt.pie(values, labels=labels, startangle=90, autopct='%.0f%%', counterclock=False)
plt.title("Kuva 3")
plt.show()
```

```python
#- tee pylväskaavio suurimmista kaupungeista, jossa yksi pylväs on väkiluku ja toinen pylväs on työpaikkojen määrä

x = np.array(suurimmatkaupungit['Alue'])
y = np.array(suurimmatkaupungit['Väkiluku, 2024'])
y2 = np.array(suurimmatkaupungit['Alueella olevia työpaikkoja, 2023'])
suurimmatkaupungit
plt.bar(x,y,label="väkiluku")
plt.bar(x,y2,label="työpaikat")

plt.xticks(rotation=90)
plt.title("Kuva 4: Työpaikkoja suhteessa väkilukuun")
plt.xlabel("kunta")
plt.ylabel("Väkiluku")
plt.legend(loc='center')
plt.show()
```
