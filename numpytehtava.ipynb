```python
#6> 
random_array = np.random.normal(size=100, loc=60, scale=0.3)
matrix6 = random_array.reshape(10,10)
matrix6
```

```python
#6.1
print(f"farvo löytyy viimeiseltä riviltä ja viimeisestä sarakkeesta: {matrix6[9][9]}")
```

```python
print(f"minimi: {np.min(matrix6)}")
print(f"maksimi: {np.max(matrix6)}")
print(f"keskiarvo: {np.mean(matrix6)}")
print(f"keskihajonta: {np.std(matrix6)}")
print(f"varianssi: {np.var(matrix6)}")
```

```python
import matplotlib.pyplot as plt
```

```python
matrix6.shape
```

```python
#plt.plot(matrix6)
plt.xlabel("Row Index")
plt.ylabel("Value")
for i in range(matrix6.shape[1]):
    plt.plot(matrix6[:, i], label=f"Column {i+1}")
plt.legend(bbox_to_anchor=(1.3, .6), loc='upper right')
plt.show()
```

```python
plt.plot(random_array)
plt.show()
```

```python
counts, bins, patches = plt.hist(matrix6.flatten(), bins=10, edgecolor='black', linewidth=1)
for c, b, p in zip(counts, bins, patches):
    x = p.get_x() + p.get_width()/4
    y = p.get_height() + .1
    plt.text(x, y,str(int(c)))
plt.xlabel("Arvo")
plt.ylabel("Frekvenssi")
plt.title("Histogrammi")
plt.show()
```
