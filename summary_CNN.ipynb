1. CNN: Convolutional Neural Network: Mạng Nơ-ron Tích Chập. Đây là một loại mạng nơ-ron đặc biệt mạnh trong xử lý dữ liệu có cấu trúc dạng lưới như:
- Ảnh (2D grid)
- Âm thanh (1D signal)
- Video (3D — theo thời gian)
- Chuỗi (sequence)

2. Tích chập là gì 
- Lấy kernel lướt trên ảnh
- Tại mỗi vị trí:
    - Nhân từng giá trị pixel với giá trị kernel tương ứng
    - Cộng tất cả lại → 1 giá trị mới
- Kết quả → feature map (ảnh đã biến đổi)

3. Kernel là gì?
- Kernel (hay filter) = ma trận nhỏ chứa các trọng số (weights), Ví dụ: 3×3, 5×5
- Mỗi giá trị trong kernel là một số học được trong quá trình training

4. Giá trị kernel đc chọn ntn?
- Khi bạn train CNN:
    - Input → convolution → feature map
    - Tính loss so với target (label)
    - Backpropagation → tính gradient của loss w.r.t. kernel weights
    - Update kernel weights (như update weights trong MLP)
- Sau nhiều epoch:
    - Kernel sẽ “tự động học” các pattern quan trọng
    - Ví dụ: Cạnh dọc / ngang, Góc, Texture
- Mỗi layer học kernel khác nhau
    - Layer đầu → học pattern đơn giản (edges, corners)
    - Layer giữa → học pattern phức tạp hơn (shapes, motifs)
    - Layer sâu → học pattern trừu tượng (mặt người, object)

5. Bn layers là đủ:
- Số layers phụ thuộc task và input size: càng nhiều layers thì càng học sâu
    - Có thể dùng 1 layer cho những task đơn giản, ví dụ: Nhận diện cạnh / số viết tay MNIST đơn giản (28x28) → 1 conv layer + pooling đủ
    - Phức tạp hơn: CIFAR-10 / ImageNet → 3–20+ conv layers
5. Channel là gì
- 1️⃣ Channel trong hình ảnh
    - Hình ảnh màu RGB = 3 channel (R = Red, G = Green, B = Blue)
    - Hình ảnh grayscale = 1 channel
    - Mỗi channel = 1 ma trận 2D lưu giá trị cường độ
    - vd: 1 matrix 4x4 lưu 1 ảnh gray 4x4 => RGB image 4×4 → 3 ma trận 4×4 → 3 channel
- 2️⃣ Channel trong CNN
    - Output của 1 convolutional layer = feature map
    - Mỗi filter / kernel tạo ra 1 channel riêng trong output
    - Nên:
        - Input image RGB → 3 channel
        - Conv layer có 8 filters → output = 8 channel (feature maps)
    - Ví dụ
        - Input: 32×32×3 (RGB image)
        - Conv layer: 3×3 kernel, 8 filters
        - Output: 30×30×8 → 8 channel, mỗi channel = 1 feature map
