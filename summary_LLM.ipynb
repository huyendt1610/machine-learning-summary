- Flow trong NLP (ví dụ Transformer)
1. Input text → Tokenizer "I love NLP" → ["I", "love", "NL", "P"]
2. Token → Embedding vector: Mỗi token map sang vector dày đặc, ví dụ [0.12, -0.34, ...]
    Các từ có nghĩa gần nhau sẽ gần nhau
3. Model xử lý:
- Neural network tính toán dựa trên embedding
- Attention, MLP… → dự đoán next token
4. Output vector → Softmax → Token
- Vector cuối cùng -> xác suất token tiếp theo
- Chọn token với max probability → sinh text
