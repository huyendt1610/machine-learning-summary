```python
import pandas as pd
import numpy as np 
```

### concat & concatenate
1. np.concatenate(axis = [0,1])
2. pd.concat(axis)

- np.stack(axis)
- np.vstack(): vertical stack 
- np.hstack(): horizontal stack 
- np.dstack(): dimension stack 

```python
a = np.linspace(2,10,5, dtype=int)
b = np.linspace(20,50,5, dtype=int)

a, b
```

```python
np.concat([a,b]) # same as np.concat([a,b], axis=None)
```

```python
a = np.array([np.linspace(2,10,5, dtype=int)])
b = np.array([np.linspace(20,50,5, dtype=int)])
```

```python
print(a)
print(b)
```

```python
np.concatenate((a,b), axis=None) 
```

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6]])
np.concatenate((a,b.T), axis=1)
```

```python
np.concatenate((a,b), axis=None)
```
