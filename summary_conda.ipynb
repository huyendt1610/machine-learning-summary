    - Cách thức download và cài đặt: Khi chạy lệnh install: -> (với OS (win, linux, mac), 32/64bit, python version) tìm theo thứ tự ưu tiên các file phù hợp
        - pip
            - **.whl** file wheel kiểu dạng binary đã được build sẵn (.pyd vs .so) -> chạy luôn ko cần build 
            - tên file: scikit_surprise-1.1.3-cp310-cp310-win_amd64.whl
                - cp310: ABI của CPython (Application Binary Interface)> chuẩn quy định các thức python code tương tác với CPython interpreter
                - cp310: python 3.10 
                - win_amd64 → Windows 64-bit

            - ko có thì sẽ download file .tar.gz là source code -> sau đó build binary cho máy local cụ thể 
            - một số thư viện có mã C++ -> cần compiler để build thành .pyd or .so -> nếu local ko có -> bị lỗi

        - conda: 
            - **.conda** đã chứa binary build sẵn, ko cần compiler để build nữa (là định dạng mới so với .tar.bz2)
            - Vì hệ thống mở conda-forge nên là hệ thống binary đã build này rất rộng ko cần tự build từ source 
            - Khi user cung cấp recipe cho conda -> tự động build các binary cho các hệ điều hành, cấu trúc, version khác nhau của python 
            - Conda đảm bảo tất cả libraries trong env dùng chung ABI tương thích -> nếu cần build lại toàn bộ 
            - tên file: scikit_surprise-1.1.3-py311h3dcd195_0.conda (hoặc .tar.bz2)
                - py311: python 3.11 
                - h3dcd195_0: Đây là build string, chỉ ra các đặc điểm của việc biên dịch thư viện (bao gồm ABI, các dependency, v.v.).

2. Anaconda + miniconda: là software distribution (là collection of packages mà có thể install trên 1 hệ thống)
   - **Anaconda**: 
        - created by Continuum Analytics (nổi tiếng trong analytics)
        - a full distribution of the centre software 
        - contain python and several hundred third party open-source projects
    - **Miniconda**: 
        - an installer for an empty conda environment, 
        - containing only conda and its dependencies -> 
        - can install from scratch 
    - **Miniforge**: (biến thể của miniconda với package channels là conda-forge)

**packages**
- conda install -c conda-forge scikit-surprise -> install với channel conda-forge package tên là scikit-surprise
- conda install -n env_name pandas -> cài đặt pandas vào môi trường [env_name] với tham số -n = --name 
- conda install pandas numpy matplotlib -> ko cần chỉ rõ channel -> lấy từ các pacakges of anaconda 
- pip install ... -> dùng chạy pip bình thường trong conda 
    - khi này pip sẽ re-compound các file lại với nhau và cài đặt
    - vì các file của pip chỉ là các file .whl hoặc build distribution 
    - nên nó cần rebuild lại tùy theo các hệ điều hành mà nó cài đặt 

**channels**
- conda config --show channels -> hiển thị all channels hiện có 
- conda config --add channels conda-forge -> add more channels 
- conda config --set channel_priority strict -> lấy package theo thứ tự từ trên xuống dưới trong channels list 
- conda config --set channel_priority flexible -> lấy ở đâu cũng đc
- conda config --show channel_priority -> hiển thị xem đang set là giá trị nào 

**environment**
- conda env list: kiểm tra xem conda có những environment nào 
- conda info --envs: tương tự như trên 

- conda create -n surprise-env python=3.11 -> tạo 1 environemtn với --name là surprise-env với python version
- conda deactivate --> deactivate current env và back lại base env 

- conda create --prefix ./env python=3.10 -> dùng prefix để tùy chỉnh đường dẫn cho folder env 
- conda config --set env_prompt '({name})' -> set đg dẫn trong prompt chỉ là (tên) thay vì path cho environment 
- conda remove --prefix ./env --all
