### Python for web development
- Framework:
    - Flask: A microframework for web apps, (Very few — just routing, request/response handling, and Jinja templates.) 
    - Django: A full-stack web framework: ORM (Object-Relational Mapper), admin, and auth (blog, shop, SaaS shop)
    - FastAPI: modern & API-first, high performace async API (no built-in ORM, admin, or auth)
- Interface: between web server and python web app
    - WSGI: Web Sever Gateway Interface (sync).
    - ASGI: Asynchronous Sever Gateway Interface (async, supports WebSockets, async IO)
<br>
- Web server: (like Nginx, Apache)
    - Gunicorn for WSGI
    - Uvicorn for ASGI

KDE: Kernel Density Estimate (biểu đồ mật độ xác suất)

- trị riêng ($\lambda$) của 1 ma trận là thoa mãn Ax =  $\lambda$.x -> giải hệ pt: (A- $\lambda$.I).x =  $\theta$ (1)
-> det(A- $\lambda$.I) = 0 -> $\lambda$
- thay vào (1) tìm đc x là vector riêng

Vector riêng có ứng dụng lớn nhất là để chéo hóa ma trận (chỉ có thể áp dụng cho ma trận vuông nxn)
- A gọi là chéo hóa đc nếu tồn tại 1 ma trận T vuông, khả nghịch và tồn tại ma trận đường chéo đc tính theo công thức:<br>
$$
T^{-1}.A.T
$$ 
- A là ma trận chéo hóa đc
- T là ma trận làm chéo hóa A:
    - T là ma trận gồm các cột là các VTR (vector riêng)
    - B = T^{-1}.A.T là ma trận A đã đc làm chéo chính là matrix với các giá trị trên đường chéo chính là các GTR (giá trị riêng của A)
<br> => $A = T.B.T^{-1}$
<br> => det(A) = det($T.B.T^{-1}$)
<br> => $A^{2021} = T.B^{2021}.T^{-1}$ (vì B là matrix đường chéo nên chỉ cần lũy thừa 2021 lần các phần tử trên đường chéo chính)
