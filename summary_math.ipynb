### Trong thực tế, data analytíc/science 
1. Loss function: hàm mất mát
- quan hệ giữa y và y*: giá trị thực tế và giá trị dự đoán 
(y* là giá trị đoán của y, f* là hàm dự đoán của f)
- mục tiêu là sai số này càng ít càng tốt -> dùng đạo hàm tìm giá trị cực tiểu của f*
và xây dựng được hàm f* phức tạp dựa trên các features (alennus, vouchers, incomes,..)
2. How
- thử với giá trị x1, nếu giá trị đạo hàm dương -> hàm đang tăng -> muốn tìm **cực tiểu** -> nhích X1 về bên trái hàm số
- nếu tại x1, đạo hàm < 0 -> hàm đang giảm -> muốn tìm **cực tiểu** -> nhích về bên phải x1 hàm số
- **muốn xác định hàm có cực tiểu hay cực đại -> hàm lồi hay lõm (sử dụng đạo hàm bậc )**
