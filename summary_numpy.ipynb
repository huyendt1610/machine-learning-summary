- np.random.rand(n): sinh ra n số trong nửa đoạn [0, 1) theo uniform distribution (phân phối đều)
- np.random.seed(n): đảm bảo nếu gieo cùng 1 seed thì kết quả trả về của rand là ko đổi 
- np.random.randn(n): sinh ra n số theo phân phối chuẩn Gaussian với kỳ vọng và phương sai là 0, 1
- np.random.normal(loc, scale, size): 
    - loc: shift distribution left or right on X axis (loc=μ)
    - scale: spread of dist (scale=σ)
- np.random.randint(from, to, size=number or tuple)

```python
import numpy as np
```

```python
np.random.randn(4).reshape(2,2)
```

```python
def generate_normal_dist(mu: float, sigma: float, m: int, n: int):
    arr = np.random.randn(m*n)
    np.random.seed(42)
    arr = arr * sigma + mu
    return arr.reshape(m,n)
```

```python
generate_normal_dist(1,1,4,2)
```

```python
np.random.normal(loc=50, scale=5, size=8) #(loc=μ, scale=σ, size=n)
```

```python
np.random.randint(2,5,((3,2)))
```

```python
np.random.permutation(10) # 1 hoán vị của các số từ 0-9
```
