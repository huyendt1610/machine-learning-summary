### Read Json file: json (built-in), pandas, requests

- json.load(from file)
- json.loads(from str with json structure)
 
- json.dump(to file)
- json.dumps(to str)

```python
with open("suurimmatkaupungit.json", "r") as file: 
    data = json.load(file)
data['Alue'] #dictionary => access by key
print(type(data))
```

```python
with open("suurimmatkaupungit.json", "r") as file: 
    data_str = file.read()

data = json.loads(data_str) #return dict
print(type(data))
```

### Write json file> json (dump, dumps), pandas (arg is a DataFrame)

```python
data = {
    "name": "Alice",
    "age": 23,
    "languages": ["English", "Finnish"]
}
```

```python
with open("writejson.json","w") as file:
    json.dump(data, file)
```

```python
json_str = json.dumps(data) #conver from data to str
with open("writejson.json","w") as file:
    file.write(json_str)
```

```python
df = pd.DataFrame(data=data)
df
df.to_json("writejson2.json", orient="records")
```
