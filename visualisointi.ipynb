```python
#pip install pandas seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.get_dataset_names() #return names of diff example dataset that we can use 
```

```python
df = sns.load_dataset('titanic') #get dataset name titanic, return DataFrame
df.head(15) 
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe() #std, min, max
```

```python
df.isna() # is N/A: missing value or not 
```

```python
df.dropna() #remove missing value 
```

```python
df.duplicated()
df = df.drop_duplicates(ignore_index=True) #if ignore_index=False, se voi change index of df
df.shape
```

```python
df.describe()
```

```python
df.hist(figsize=(10,7), bins=20)
plt.show()
```

```python
sns.boxplot(data=df, x='fare')
plt.show()
```

```python
plt.figure(figsize=(7,5)) #default value (5,5) inchs (width, height)

sns.boxplot(data=df, x='pclass', y='fare', hue='survived')
plt.title('Matkalipun hinta ja matkustusluokka suhteessa selviytymiseen')
#plt.ylim(0,300)

plt.show()
```

```python
daa = df[(df["pclass"] ==1) & (df["who"]=="man")]
daa.describe() #barplot draw the mean value of each column
```

```python
#sns.barplot(data=df, x='survived', y='who', errorbar=None)
sns.barplot(data=df, x='pclass', y='survived', hue='who', palette='pastel') #hue, palette: color theme
plt.show()
```

```python
df['fare_arange'] = pd.qcut(df['fare'],4) #separate into 4 groups 
df
```

```python
sns.barplot(data=df, x='survived', y='fare_arange')
plt.show()
```

```python
df2 = df.groupby(['sex','alive']).size().reset_index(name='number')
df2
```

```python
sns.barplot(data=df2, x='sex', y='number', hue='alive')
plt.show()
```

```python
df3 = df.drop(['sex', 'embarked', 'class', 'who', 'embark_town', 'deck', 'alive', 'fare_arange'], axis=1)
df3
```

```python
corr_matrix = df3.corr() #correlation 
corr_matrix
```

```python
sns.heatmap(corr_matrix, vmin=-1.0, vmax=1.0, cmap='RdBu', annot=True) #annot: present the value of cell on it
```
